```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True) # we are deleting all the NaN values for a smooth dataset
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split # Splitting the data into training and testing data

X = data.drop(['median_house_value'], axis = 1) # x is the input, hence we drop the target variable
y = data['median_house_value'] # y will be the target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)  # Specifying the split percentage
```

```python
train_data = X_train.join(y_train) # Combining the I/O for the training data
```

```python
train_data
```

```python
train_data.hist(figsize = (15, 8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
# Data Pre-Processing

#In order to normalize the left-skewed graphs, take the log and add 1 to prevent any 0 values

train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# We have an feature that is non-numerical, namely, ocean proximity

train_data.ocean_proximity.value_counts()
```

```python
# Since it affects the target, we turn each of its types into a feature and then give it a boolean value 

train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
train_data = train_data.drop(['ISLAND'], axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
# Visualizing the relation between latitudes/longitudes and the median house value

plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='coolwarm')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['pop_per_hhold'] = train_data['population']/train_data['households']
train_data['rooms_per_pop'] = train_data['total_rooms']/train_data['population']
train_data['bedrooms_per_pop'] = train_data['total_bedrooms']/train_data['population']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)


reg = LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
test_data = X_test.join(y_test) # Combining the I/O for the testing data
```

```python
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)


test_data['pop_per_hhold'] = test_data['population']/test_data['households']
test_data['rooms_per_pop'] = test_data['total_rooms']/test_data['population']
test_data['bedrooms_per_pop'] = test_data['total_bedrooms']/test_data['population']
```

```python
test_data = test_data.drop(['ISLAND'], axis=1)
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test) 
```

```python
reg.score(X_test_s, y_test)  # Scaling did not change the performance much
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test) # Better than Linear Regression
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, 
                           scoring="neg_mean_squared_error",
                           return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, y_test)
```
